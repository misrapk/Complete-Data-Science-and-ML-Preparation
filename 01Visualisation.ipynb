```python
import matplotlib.pyplot as plt
```

```python
years = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
gdp = [6.39, 7.41, 8.00, 8.26, 6.80, 6.45, 3.74, -6.60, 8.68, 7, 6.8]
```

```python
plt.plot(years, gdp, color="blue", marker="o", linestyle="solid")
plt.title("India's GDP Growth last 10 years")
plt.ylabel("GDP Growth in %")
plt.xlabel("Years")

plt.show()
```

```python
#bar
plt.bar(years, gdp)
```
