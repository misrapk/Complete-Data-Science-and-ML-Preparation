```python
import nltk
import re
```

```python
import pandas as pd
msgs = pd.read_csv('data/SMSSpamCollection.txt', sep='\t', names=['label', 'messages'])

msgs
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#clean the data
corpus = []
for i in range(0, len(msgs)):
    review = re.sub('[^a-zA-Z]', ' ', msgs['messages'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
len(corpus)
```

# 1.  Using BOW

```python
from sklearn.feature_extraction.text import CountVectorizer #create BOW model
cv = CountVectorizer(max_features=2500, ngram_range=(1,2))   #for binary BOW, enable binary= True
```

```python
X = cv.fit_transform(corpus).toarray()  #independent features
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=10000, formatter=dict(float=lambda x: "%.3g" %x))
X
```

```python
cv.vocabulary_
```

```python
#output feature

y= pd.get_dummies(msgs['label'], dtype=int)
y
```

```python
y = y.iloc[:, 0].values
```

```python
# train test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train ,y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Apply alog

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model = MultinomialNB()
spam_detect_model.fit(X_train, y_train)
```

```python
y_pred = spam_detect_model.predict(X_test)
y_pred
```

```python
#accuracy
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
