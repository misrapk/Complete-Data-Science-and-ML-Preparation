```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#create datapoints
from sklearn.datasets import make_classification

X,y= make_classification(n_samples=1000, n_features=2, n_classes=2,n_clusters_per_class=1,n_redundant=0)
```

```python
X,y
```

```python
#scatter plot
sns.scatterplot( x= pd.DataFrame(X)[1],y = pd.DataFrame(X)[0],hue=y)
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
#accuracy metriccecs
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
y_pred = svc.predict(X_test)
print("Accuracy Score",accuracy_score(y_test,y_pred))
print("Confusoin Matrix: \n", confusion_matrix(y_test,y_pred))
print("Classification Report: \n",classification_report(y_test,y_pred))
```

```python
#draw the svm line and boundry
w = svc.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-3,3)
yy = a * xx - (svc.intercept_[0]) / w[1]

sns.scatterplot( x= pd.DataFrame(X)[1],y = pd.DataFrame(X)[0],hue=y)
plt.plot(xx,yy)
plt.show()
```

# Hyper paramter tuning with SVC

```python
from sklearn.model_selection import GridSearchCV

#parameters
param_grid = {'C':[0.1,1,10,100,1000],
              'gamma':[1,0.1,0.01,0.001,0.0001],
              'kernel':['rbf', 'poly', 'sigmoid','linear']}
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=3)
```

```python
grid.fit(X2_train,y2_train)
```

```python
grid.best_params_
```

```python
#prediction
grid_pred = grid.predict(X2_test)
print("Accuracy Score",accuracy_score(y2_test,grid_pred))
print("Confusoin Matrix: \n", confusion_matrix(y2_test,grid_pred))
print("Classification Report: \n",classification_report(y2_test,grid_pred))
```

```python
#LETS DRAW THE RBF ON THE GRAPH WITH GRIDCV PARAM
best_estimator = grid.best_estimator_

if hasattr(best_estimator, 'coef_'):
	w = best_estimator.coef_[0]
	a = -w[0] / w[1]
	xx = np.linspace(-3, 3)
	yy = a * xx - (best_estimator.intercept_[0]) / w[1]

	sns.scatterplot(x=pd.DataFrame(X2)[1], y=pd.DataFrame(X2)[0], hue=y2)
	plt.plot(xx, yy)
	plt.show()
else:
	print("The best estimator does not have a linear kernel, so the decision boundary cannot be plotted.")

```
